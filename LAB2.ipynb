```python
import numpy as np
```

```python
dataFile = open('instances_labs/binpack.txt','r')

box_size = int(dataFile.readline().split('=')[1])
data = []

for i in dataFile.readlines():
    data.append(float(i))

data = np.array(data)

dataFile.close()
np.random.shuffle(data)
print(data)
print(box_size)
```

# Online

## Next-fit(NF) ##

```python
def NF(data, size):
    box_size = size
    items = data
    box = 1
    ost = box_size
    
    for item in items:
        if item > ost+1e-5: 
            box += 1
            ost = box_size - item
        else:
            ost -= item
    return box
```

```python
NF(data, box_size)
```

## First-fit(FF) ##

```python
def FF(data, size):
    box_size = size
    items = data
    box_list = [0]
    
    for item in items:
        for i, size in enumerate(box_list):
            if size + item <= box_size+1e-5:
                box_list[i] += item
                break
        else:
            box_list.append(item)
    print(box_list)
    return len(box_list)
```

```python
FF(data, box_size)
```

# Offline #

## Next-fit-decutresreducing (NFD) ##

```python
def NFD(data, size):
    box_size = size
    items = data
    sorted_items = sorted(items, reverse=True)
    return NF(sorted_items, box_size)
```

```python
NFD(data, box_size)
```

## First-fit-decreasing(FFD) ## 

```python
def FFD(data, size):
    box_size = size
    items = data
    sorted_items = sorted(items, reverse=True)
    return FF(sorted_items, box_size)
```

```python
FFD(data, box_size)
```
