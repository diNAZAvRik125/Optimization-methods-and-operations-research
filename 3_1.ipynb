```python
from scipy.optimize import linprog
```

```python
# Коэффициенты целевой функции (коэффициенты для минимизации)
c = [-3, -2]  # Минимизируем -3x - 2y, чтобы максимизировать количество рам
```

```python
# Коэффициенты ограничений (в виде матрицы A и вектора b)
A = [
    [1, 0],    # x <= 100
    [0, 1],    # y <= 200
    [-3, -2]  # 3x + 1.5y <= 100 + 200 + 300
]
```

```python
b = [100, 200, 300]
```

```python
# Границы переменных (x >= 0, y >= 0)
x_bounds = (0, None)
y_bounds = (0, None)
```

```python
# Решение задачи линейного программирования
result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')
```

```python
print("Результаты распила:")
print("Количество метровых заготовок (x):", result.x[0])
print("Количество полутораметровых заготовок (y):", result.x[1])
print("Максимальное количество рам:", int(-result.fun))  # Минимизация, поэтому инвертируем знак
```
