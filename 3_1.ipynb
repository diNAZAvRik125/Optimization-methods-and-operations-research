```python
from scipy.optimize import linprog
```

```python
#    (  )
c = [-3, -2]  #  -3x - 2y,    
```

```python
#   (   A   b)
A = [
    [1, 0],    # x <= 100
    [0, 1],    # y <= 200
    [-3, -2]  # 3x + 1.5y <= 100 + 200 + 300
]
```

```python
b = [100, 200, 300]
```

```python
#   (x >= 0, y >= 0)
x_bounds = (0, None)
y_bounds = (0, None)
```

```python
#    
result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')
```

```python
print(" :")
print("   (x):", result.x[0])
print("   (y):", result.x[1])
print("  :", int(-result.fun))  # ,   
```
